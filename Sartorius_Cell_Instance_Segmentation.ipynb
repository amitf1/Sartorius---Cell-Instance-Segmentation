### Data Visualization and EDA

```python
data = glob.glob(os.path.join(data_path, '*.png'))
```

```python
def show_rand_n_images(data, n, rows, cols):
  rand_images = np.random.choice(data, n)
  fig, ax = plt.subplots(rows,cols, figsize=(20,15))
  ax = ax.ravel()
  for i, image_path in enumerate(rand_images):
    img = cv2.imread(image_path)
    ax[i].set_title(os.path.basename(image_path))
    ax[i].axis("off")
    ax[i].imshow(img);
```

```python
show_rand_n_images(data, 4, 2, 2)
```
